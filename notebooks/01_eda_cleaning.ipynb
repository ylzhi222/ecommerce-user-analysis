```python
#å¯¼å…¥å¸¸ç”¨åŒ…
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import gzip

#é…ç½®ç»˜å›¾é£æ ¼
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

#è®¾ç½®æ•°æ®è·¯å¾„
DATA_DIR = "../data/"
OUTPUT_DIR = "../output/"
FILE_NAME = "2019-Oct.csv"

#1. è¯»å–æ•°æ®
file_path = os.path.join(DATA_DIR, FILE_NAME)

#æ£€æŸ¥æ–‡ä»¶æ˜¯å¦å­˜åœ¨
if not os.path.exists(file_path):
    print(f"âš ï¸ æ–‡ä»¶æœªæ‰¾åˆ°: {file_path}. åˆ›å»ºä¸€ä¸ªå ä½ç¬¦æ–‡ä»¶ç”¨äºæµ‹è¯•ã€‚")
    os.makedirs(DATA_DIR, exist_ok=True)
    # åˆ›å»ºä¸€ä¸ªç©ºçš„å ä½ç¬¦æ–‡ä»¶
    with open(file_path, 'w') as f:
        f.write("event_time,event_type,category_code,brand,price\n")

#æ£€æµ‹æ–‡ä»¶æ˜¯å¦ä¸ºgzipæ ¼å¼
try:
    with gzip.open(file_path, 'rb') as gz_file:
        gz_file.read(1)  # å°è¯•è¯»å–ç¬¬ä¸€ä¸ªå­—èŠ‚
    is_gzip = True
except OSError:
    is_gzip = False

#æ ¹æ®æ–‡ä»¶æ ¼å¼è¯»å–æ•°æ®
if is_gzip:
    df = pd.read_csv(file_path, compression='gzip')
else:
    print("âš ï¸ æ–‡ä»¶ä¸æ˜¯gzipæ ¼å¼ï¼ŒæŒ‰æ™®é€šCSVæ–‡ä»¶è¯»å–ã€‚")
    df = pd.read_csv(file_path)

print(f"âœ… æ•°æ®è¯»å–å®Œæˆï¼Œå¤§å°ä¸º: {df.shape}")

#2. æŸ¥çœ‹åŸºæœ¬ä¿¡æ¯
print("\nğŸ“‘ åˆ—åï¼š")
print(df.columns.tolist())

print("\nğŸ” æ•°æ®ç±»å‹ï¼š")
print(df.dtypes)

print("\nğŸ‘€ å‰5è¡Œé¢„è§ˆï¼š")
print(df.head())

#3. è½¬æ¢æ—¶é—´åˆ—
df['event_time'] = pd.to_datetime(df['event_time'], errors='coerce')

#4. æ£€æŸ¥ç¼ºå¤±å€¼
print("\nâ“ ç¼ºå¤±å€¼ç»Ÿè®¡ï¼š")
print(df.isnull().sum())

#5. äº‹ä»¶ç±»å‹åˆ†å¸ƒ
print("\nğŸ›ï¸ äº‹ä»¶ç±»å‹åˆ†å¸ƒï¼š")
print(df['event_type'].value_counts())

#6. ç±»ç›®ï¼ˆcategory_codeï¼‰Top10
print("\nğŸ“¦ çƒ­é—¨ç±»åˆ«Top10ï¼š")
print(df['category_code'].value_counts().head(10))

#7. å“ç‰Œï¼ˆbrandï¼‰Top10
print("\nğŸ·ï¸ çƒ­é—¨å“ç‰ŒTop10ï¼š")
print(df['brand'].value_counts().head(10))

#8. ä¿å­˜ä¸€ä¸ªå°æ ·æœ¬ï¼ˆ10ä¸‡æ¡ï¼‰ç”¨äºå¿«é€Ÿè°ƒè¯•
sample_df = df.sample(100000, random_state=42)

# ç¡®ä¿outputç›®å½•å­˜åœ¨
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ä¿å­˜æ ·æœ¬
sample_file = os.path.join(OUTPUT_DIR, "cleaned_data_sample.csv")
sample_df.to_csv(sample_file, index=False)

print(f"\nâœ… æ ·æœ¬ä¿å­˜æˆåŠŸ: {sample_file}")
```
